## Data Visualisation

```python
plt.figure(figsize=(20,10))
sns.countplot('Ship Mode',data=df)
plt.title('Counts by Ship Mode',fontsize=20)
plt.show()
```

```python
category_profit=df.groupby('Category')['Profit'].sum()
category_profit
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=category_profit.index,y=category_profit.values)
plt.title('Category wise profit')
plt.show()
```

## Conclusion:

1. Sales is directly proportional to profit.AS sales increases profit also increases.
2. Discount is inversely proportional to profit.Customer should be given less discount.
3. In 'Sub-Category' tables, bookcases and supplies, profit are less.So,this sub-category should be focused to increase profit.
4. Southern and Central region customers should be focused to increase the profit.
5. Quantity doesn't decide profit.
6. Furniture products should be focused to increase profit.
